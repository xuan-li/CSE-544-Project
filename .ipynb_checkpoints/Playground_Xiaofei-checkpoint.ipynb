```python
import re
from tqdm import tqdm_notebook as tqdm
```

```python
resourece_suffix={'jpg','png'}
```

```python
p = re.compile('http://en.wikipedia.org/wiki/[a-z,A-Z,_]*')
```

```python
# a='https://es.wikipedia.org/favicon'
a='http://en.wikipedia.org/wiki/english'
```

```python
print(p.match(a))
```

```python
import codecs
names=[]
file_name='./data/wiki.1190153705'
with codecs.open(file_name, "r",encoding='utf-8', errors='ignore') as fdata:
    for line in tqdm(fdata):
        line=line.strip().split(' ')
        url=line[2]
        if p.match(url):
            name=url.split('/')[-1]
            names.append(name)
```

```python
len(names)
```

```python
from collections import Counter
c=Counter(names)
```

```python
c=list(c.items())
```

```python
sorted(c,key=lambda x:x[1], reverse=True)
```
