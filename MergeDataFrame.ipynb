```python
import numpy as np
import os
import threading
import operator
import pickle
import re
import pandas as pd


start_time = 1190146243
end_time = 1192994591

def dump_data(data, output_name):
    with open(output_name, 'wb') as f:
        pickle.dump(data, f)

def load_data(input_name):
    with open(input_name, 'rb') as f:
        data = pickle.load(f)
    return data
```

```python
from urllib.parse import unquote
from urllib.parse import quote

def unquote_word(word):
    word = " ".join(unquote(unquote(word)).split("_"))
    return word

word_freq = load_data('data/dump/hot_words_100000.pkl')
unquoted_to_quote = {unquote_word(word): word for word in word_freq.index}
dump_data(unquoted_to_quote, 'data/dump/unquoted_to_quoted.pkl')
```

```python
metadata1 = load_data("data/dump/meta_data_num_info.pkl")
metadata2 = load_data("data/dump/access_update_dataframe.pkl")
vandalism = load_data("data/dump/access_vandalism_dataframe.pkl")
graph = load_data("data/dump/graph_info.pkl")
pagelen = pd.DataFrame.from_csv("data/dump/visit_count_vs_page_length.csv")
```

```python
graph = graph.rename(unquoted_to_quote, axis=0)
```

```python
merge = pd.concat([metadata1, metadata2, vandalism, graph, pagelen], axis = 1)
```

```python
cleaned_merged = merge.drop(['Visit Count','access', 'freq','Entry'],axis=1).dropna(axis=0, how='any')
```

```python
dump_data(cleaned_merged, 'data/dump/merged_metadata.pkl')
```
